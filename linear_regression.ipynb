```python
# MultiRegression 
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
# load the data
df=sns.load_dataset('tips')
df['smoker_num']=df['smoker'].map({'Yes':1, "No":0})
# select the feature column from df
X=df[['total_bill', 'size', 'smoker_num']]
#select the target column
y=df['tip']
#load/call the linear regression modal and fit the model
model=LinearRegression()
model.fit(X, y)
```

```python
print(model.predict([[40.223, 5, 1]]))
print(model.predict([[35.223, 2, 0]]))
```
