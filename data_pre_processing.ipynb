```python
import pandas as pd
import numpy as np
import seaborn as sns


# load the Titalnic Data set
data = sns.load_dataset("titanic")
data.head()
```

```python
# check the number missing values in colunns
data.isnull().sum().sort_values(ascending=False)
```

```python
# drop deck colunn
# data.drop('deck', axis=1, inplace=True)
data = data.drop("deck", axis=1)
data.head()
```

```python
data["age"] = data["age"].fillna(data["age"].mean())
data.isnull().sum().sort_values(ascending=False)
```

```python
# impute missing values with mode
data["embark_town"] = data["embark_town"].fillna(data["embark_town"].mode()[0])
data["embarked"] = data["embarked"].fillna(data["embarked"].mode()[0])

data.isnull().sum().sort_values(ascending=False)
```

## K-Nearest Neighbors(KNN)


```python
data.isnull().sum().sort_values(ascending=False)
```

```python
# imute missign values with KNN imputer
from sklearn.impute import KNNImputer

# call the KNN class with number of neighbors
imputer = KNNImputer(n_neighbors=4)

# impute the missing values with KNN imputer


data["age"] = imputer.fit_transform(data[["age"]])


data.isnull().sum().sort_values(ascending=False)
```

```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# load the dataset
df = sns.load_dataset("titanic")
df.head()
```

```python
imputer = IterativeImputer(max_iter=10)

columns_to_impute = ["age", "embark_town", "embarked", "deck"]
for col in columns_to_impute:
    df[col] = imputer.fit_transform(df[[col]])

df.isnull().sum().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import LabelEncoder

columns_to_encode = ["sex", "embarked", "who", "deck", "class", "embark_town", "alive"]
label_encoders = {}

for col in columns_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

    label_encoders[col] = le

df.head()
```

## Inverse Transform


```python
# inverse transform for encoded columns
for col in columns_to_encode:
    le = label_encoders[col]
    df[col] = le.inverse_transform(df[col].astype(int))

df.head()
```

### Random Forest for Impute values


```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    mean_squared_error,
    r2_score,
    mean_absolute_error,
    mean_absolute_percentage_error,
)

# load the dataset
df = sns.load_dataset("titanic")
df.head()
df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop("deck", axis=1, inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import LabelEncoder

columns_to_encode = ["sex", "embarked", "who", "class", "embark_town", "alive"]
label_encoders = {}

for col in columns_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

    label_encoders[col] = le

df.head()
```

```python
# split the dataset into two parts: one with missing values, one without
df_with_missing = df[df["age"].isna()]

# dropna removes all rows with missign values
df_without_missing = df.dropna()
```

```python
print("Original dataset", df.shape)
print("with Missing dataset", df_with_missing.shape)
print("without dataset", df_without_missing.shape)
```

```python
print(df.columns)
```

```python
# split the data into X and y and we will only take the column swith no missign values
X = df_without_missing.drop(["age"], axis=1)
y = df_without_missing["age"]


# split the data into tran and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# RandomForset Imputation
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
```

```python
print(
    f"RMSE for random forest imputation, {np.sqrt(mean_squared_error(y_test, y_pred))}"
)  # evalute teh model
print(
    f"R2 Score for random forest imputation, {r2_score(y_test, y_pred)}"
)  # evalute teh model
print(
    f"MAE for random forest imputation, {mean_absolute_error(y_test, y_pred)}"
)  # evalute teh model
print(
    f"MAPE         for random forest imputation, {mean_absolute_percentage_error(y_test, y_pred)}"
)  # evalute teh model
```

```python
df_with_missing.isnull().sum().sort_values(ascending=False)
```

```python
y_pred=rf_model.predict(df_with_missing.drop(['age'], axis=1))
y_pred
```

```python
df_with_missing['age']=y_pred
df_with_missing.isnull().sum().sort_values(ascending=False)
```

```python
# concatenate two dateframe
df_complete=pd.concat([df_with_missing, df_without_missing], axis=0)

print("Shape", df_complete.shape)
df_complete.head()
```
